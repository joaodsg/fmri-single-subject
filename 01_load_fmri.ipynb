```python
from nilearn import image, plotting

# fMRI file you uploaded
fmri_file = "sub-17017_task-rest_bold.nii"

# Load the image
img = image.load_img(fmri_file)

# Compute mean image across time
mean_img = image.mean_img(img)

# Plot the mean EPI image
plotting.plot_epi(mean_img, title="Mean fMRI Image")
```
